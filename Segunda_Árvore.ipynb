```python
import pandas as pd
from sklearn import tree
from sklearn import metrics
from sklearn.model_selection import train_test_split #nova biblioteca
import matplotlib.pyplot as plt
```

```python
treino=pd.DataFrame ({
    "Dorpeito":["N","N","S","S","S","N"],
    "peso":[90,70,80,70,90,80],
    "ataquecard":["S","N","S","S","S","N"],
})
print(treino)
```

```python
xtreino=treino.iloc[:,:-1]
ytreino=treino.iloc[:,-1]
print(xtreino)
print(" ") #separar
print(ytreino)
```

```python
xtreino=xtreino.replace({"N":0,"S":1})
model=tree.DecisionTreeClassifier(criterion="entropy")
model.fit(xtreino,ytreino)
```

```python
plt.show(tree.plot_tree(model))
```

```python
teste=pd.DataFrame ({
    "Dorpeito":["N","N","S","S"],
    "peso":[65,85,75,50],
    "ataquecard":["N","N","S","S"],
})
print(teste)
```

```python
xteste=teste.iloc[:,:-1]
yteste=teste.iloc[:,-1]

xteste=xteste.replace({"N":0,"S":1})
y_pred=model.predict(xteste)
print(f"avaliação da árvore",y_pred)
```

```python
print(metrics.confusion_matrix(yteste,y_pred))
```
