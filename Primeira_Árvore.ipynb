```python
import pandas as pd
from sklearn import tree
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
treino=pd.DataFrame ({
    "salario":["alto","baixo","baixo","alto","alto","baixo"],
    "localização":["longe","perto","longe","longe","perto","perto"],
    "função":["interess","desinteress","interess","desinteress","interess","desinteress"],
    "decisão":["sim","não","sim","não","sim","não"]
})
print(treino)
```

```python
xtreino=treino.iloc[:,:-1]
ytreino=treino.iloc[:,-1]
print(xtreino)
print(" ") #separar
print(ytreino)
```

```python
alto=1
baixo=0
perto=1
longe=0
interess=1
desinteress=0

xtreino=xtreino.replace({"alto":alto,"baixo":baixo,"perto":perto,"longe":longe,"interess":interess,"desinteress":desinteress})
```

```python
model=tree.DecisionTreeClassifier(criterion="entropy")
model.fit(xtreino,ytreino)
```

```python
plt.show(tree.plot_tree(model))
```
